## Jointplot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
## kind= reg

sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

## Pair plot
#### Also known as scatterplot. 
#### Multivariate Analysis
#### Auto pick Int or Flot variable


```python
sns.pairplot(df)
```

```python
## Add categorical variable in it.
```

```python
sns.pairplot(df, hue='sex')
```

## Distplot
### Its shows a histogram.
### use in one variable analysis

```python
sns.distplot(df['tip'])  ## y axis shows density in percentage 
```

```python
sns.distplot(df['tip'], kde=False, bins=20)
```

# Categorical Plots

```python
# Boxplot
# viollinplot
# countplot
# barplot
```

```python
df.head()
```

## Countplot
#### Show the counts of observations in each categorical bin using bars.
#### Either give x or y axis. One axis always shows the count.

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='day', data=df)
```

## Barplot
#### Show point estimates and confidence intervals as rectangular bars.
####  give both x and y axis.

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

## Boxplot
#### Percentile Chat, provide both x and y axis.
#### One Categorical and One Numeric variable req

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot('smoker', 'tip', data=df)
```

```python
sns.boxplot('day', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')  ## Show all the variable 
```

```python
sns.boxplot('total_bill', 'day', hue='smoker', data=df)
```

## Violinplot
#### Draw a combination of boxplot and kernel density estimate.

```python
sns.violinplot('total_bill', 'day', data=df , palette='rainbow')
```
