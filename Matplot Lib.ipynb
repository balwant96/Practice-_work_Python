# MatplotLib

## Its a visulazation library use for plotting. 

### import matplotLib

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Scatter plot
###  It uses dots to represent values for two different numeric variables.

```python
# plotting using matplotlib

##plt scatter   

plt.scatter(x,y,c='r')   # c- colour , g-green, b-blue, r-red
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D graph')

plt.savefig('plt_test.jpg')   # save chat as png can use jpg also
```

# plt plot
## draw Connected line 

```python
y=x*x
```

```python
plt.plot(x,y,'b*-',linestyle='dashed', linewidth=2, markersize=9)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('plt graph')
```

## subplot

```python
plt.subplot(2,3,1)    ##(2,3,1) refers 2 rows, 3 column, position 1
plt.plot(x,y,'b')
plt.subplot(1,3,3)
plt.plot(x,y,'ro--')
plt.subplot(2,3,5)
plt.plot(x,x,'g+-')
```

## Bargraph

```python
x= [3,6,4,7]
y= [7,3,5,4.5]

x2=[4,6,7,2]
y2=[2,9,4,8]

plt.bar(x,y, color='y')
plt.bar(x2,y2, color='g')
```

## Histograms

```python
a = np.array([22,5,8,55,64,34,78,45,12,23,67,43,78,98,2,65,34,56])
plt.hist(a)
plt.title('Histogram')
```

## Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]


plt.boxplot(data, vert=True, patch_artist=True)  ##vert = vertical,   patch_artist= color the data
plt.title('Box Plot')
```

## Pie 

```python
labels ='python','R','SQL','VBA'
sizes= [265,186,130,60]

explode = (0.1,0,0,0)   # explode 1st slice


#Plot
plt.pie(sizes, explode= explode, labels= labels, shadow=True, autopct='%1.1f%%')

plt.title('Pie')
```
